```python
# ==========================================
# Import Required Libraries
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ==========================================
# Load Final Cleaned Dataset
# ==========================================

df_final = pd.read_csv("../data/vehicle_final_dataset.csv")

# Display first five rows
df_final.head()
```

```python
# ==========================================
# Check Dataset Shape
# ==========================================

print("Rows :", df_final.shape[0])
print("Columns :", df_final.shape[1])
```

```python
# ==========================================
# Brand vs Fuel Type
# ==========================================

brand_fuel = pd.crosstab(
    df_final["oem"],
    df_final["fuel"]
)

brand_fuel.head()
```

```python
# ==========================================
# Average Price by Brand
# ==========================================

avg_brand_price = (
    df_final
    .groupby("oem")["listed_price"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

plt.bar(
    avg_brand_price.head(15).index,
    avg_brand_price.head(15).values
)

plt.xticks(rotation=45)

plt.title("Average Vehicle Price by Brand")

plt.xlabel("Brand")

plt.ylabel("Average Price")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Average Price by Fuel Type
# ==========================================

avg_fuel_price = (
    df_final
    .groupby("fuel")["listed_price"]
    .mean()
)

plt.figure(figsize=(8,5))

plt.bar(
    avg_fuel_price.index,
    avg_fuel_price.values
)

plt.title("Average Price by Fuel Type")

plt.xlabel("Fuel Type")

plt.ylabel("Average Price")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Average Price by Transmission
# ==========================================

avg_trans_price = (
    df_final
    .groupby("transmission")["listed_price"]
    .mean()
)

plt.figure(figsize=(6,5))

plt.bar(
    avg_trans_price.index,
    avg_trans_price.values
)

plt.title("Average Price by Transmission")

plt.xlabel("Transmission")

plt.ylabel("Average Price")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Brand vs Average Kilometer Driven
# ==========================================

avg_km = (
    df_final
    .groupby("oem")["km"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

plt.bar(
    avg_km.head(15).index,
    avg_km.head(15).values
)

plt.xticks(rotation=45)

plt.title("Average Kilometer Driven by Brand")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Manufacturing Year vs Average Price
# ==========================================

year_price = (
    df_final
    .groupby("myear")["listed_price"]
    .mean()
)

plt.figure(figsize=(12,6))

plt.plot(
    year_price.index,
    year_price.values,
    marker="o"
)

plt.title("Manufacturing Year vs Average Price")

plt.xlabel("Year")

plt.ylabel("Average Price")

plt.grid(True)

plt.show()
```
