```python
#avg price by transmission
avg_trans_price = (
    df_clean.groupby("transmission")["listed_price"]
    .mean()
)

plt.figure(figsize=(6,4))

plt.bar(
    avg_trans_price.index,
    avg_trans_price.values
)

plt.title("Average Price by Transmission")

plt.xlabel("Transmission")

plt.ylabel("Average Price")

plt.show()
```
