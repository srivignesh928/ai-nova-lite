```python
# ============================================================
# Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import (
    train_test_split,
    cross_val_score,
    RandomizedSearchCV
)

from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder

from sklearn.pipeline import Pipeline

from xgboost import XGBRegressor
```

```python
# ============================================================
# Load Feature Engineered Dataset
# ============================================================

df = pd.read_csv("../data/vehicle_feature_engineered.csv")

print(df.shape)

df.head()
```

```python
# ============================================================
# Define Features and Target
# ============================================================

X = df.drop(
    columns=[
        "listed_price",
        "price_category"
    ]
)

y = df["listed_price"]
```

```python
# ============================================================
# Define Feature Types
# ============================================================

categorical_features = [

    "oem",

    "model",

    "variant",

    "fuel",

    "transmission",

    "body",

    "owner_type",

    "City",

    "state"

]

numerical_features = [

    "km",

    "car_age",

    "premium_brand"

]

preprocessor = ColumnTransformer(

    transformers=[

        (

            "categorical",

            OneHotEncoder(

                handle_unknown="ignore"

            ),

            categorical_features

        ),

        (

            "numerical",

            "passthrough",

            numerical_features

        )

    ]

)
```

```python
# ============================================================
# Baseline Pipeline
# ============================================================
# cross validation
baseline_pipeline = Pipeline(

    steps=[

        (

            "preprocessor",

            preprocessor

        ),

        (

            "model",

            XGBRegressor(

                objective="reg:squarederror",

                n_estimators=300,

                max_depth=8,

                learning_rate=0.05,

                random_state=42

            )

        )

    ]

)
```

```python
# ============================================================
# 5-Fold Cross Validation
# ============================================================

cv_scores = cross_val_score(

    baseline_pipeline,

    X,

    y,

    cv=5,

    scoring="r2",

    n_jobs=-1

)

print("=" * 60)

print("CROSS VALIDATION RESULTS")

print("=" * 60)

print()

print("Fold Scores :")

print(cv_scores)

print()

print("Average R² :", round(cv_scores.mean(),4))

print("Standard Deviation :", round(cv_scores.std(),4))
```

```python
# =====================================================
# STEP : Lightweight Cross Validation
# =====================================================

light_model = Pipeline(

    steps=[

        ("preprocessor", preprocessor),

        ("model",

            XGBRegressor(

                objective="reg:squarederror",

                n_estimators=150,

                max_depth=8,

                learning_rate=0.05,

                random_state=42,

                n_jobs=1

            )

        )

    ]

)

cv_scores = cross_val_score(

    light_model,

    X,

    y,

    cv=3,

    scoring="r2",

    n_jobs=1

)

print("Fold Scores :", cv_scores)

print("Average R² :", cv_scores.mean())

print("Standard Deviation :", cv_scores.std())
```
